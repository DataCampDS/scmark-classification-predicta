```python
variances = X_array.var(axis=0)
X_filter = X_array[:, variances>0.05]
X_filter.shape
```

```python
def remove_highly_correlated_columns(A, seuil=0.95):
    corr = np.corrcoef(A, rowvar=False)
    keep = []
    
    for i in range(A.shape[1]):
        # On garde la colonne seulement si elle n'est pas trop corrélée
        if all(abs(corr[i, j]) < seuil for j in keep):
            keep.append(i)
    
    return A[:, keep]

X_filter = remove_highly_correlated_columns(X_filter)
X_filter.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca_fit = pca.fit_transform(X_filter)
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Rate of Variance")
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlim((0,200))
plt.show()
```
